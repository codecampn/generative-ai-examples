# Init 🐸TTS Model

````python
import os
import torch
from TTS.api import TTS
import time

# Get device
device = "cuda" if torch.cuda.is_available() else "cpu"

# List available 🐸TTS models
print(TTS().list_models())

# Init TTS
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)

def delete_files_in_folder(folder_path):
    try:
        # Überprüfe, ob der Ordner existiert
        if os.path.exists(folder_path):
            # Überprüfe, ob der Ordner nicht leer ist
            if os.listdir(folder_path):
                # Lösche alle Dateien im Ordner
                for file_name in os.listdir(folder_path):
                    file_path = os.path.join(folder_path, file_name)
                    try:
                        if os.path.isfile(file_path):
                            os.unlink(file_path)
                    except Exception as e:
                        print(f"Fehler beim Löschen der Datei {file_path}: {e}")
                print(f"Alle Dateien im Ordner {folder_path} wurden gelöscht.")
            else:
                print(f"Der Ordner {folder_path} ist bereits leer.")
        else:
            print(f"Der Ordner {folder_path} existiert nicht.")
    except Exception as e:
        print(f"Fehler beim Löschen der Dateien im Ordner {folder_path}: {e}")

def create_folder(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder '{folder_path}' created.")
    else:
        print(f"Folder '{folder_path}' already exists.")

````

## Synthesize Podcast

````python
import os

speaker_one = "example_1.wav"
speaker_two = "example_2.wav"

folder_path = "out/"
create_folder(folder_path)

#clear output folder
delete_files_in_folder("out/")

cc = 0
first_person_name = ""
for item in podcast:
  print(f"Create file {cc}/{len(podcast)-1}")
  if cc == 0:
    first_person_name = item["speaker"]
  split_sentences_bool = False
  split_text = len(item["message"])
  if split_text > 250:
    split_sentences_bool = True
  else:
    split_sentences_bool = False

  if item["speaker"] == first_person_name:
    path = f"out/{cc}_output.wav"
    tts.tts_to_file( emotion="Happy", split_sentences=split_sentences_bool, text=item["message"], speaker_wav=speaker_one, language="de", file_path=path)
  else:
    path = f"out/{cc}_output.wav"
    tts.tts_to_file( emotion="Happy",split_sentences=split_sentences_bool, text=item["message"], speaker_wav=speaker_two, language="de", file_path=path)
  cc += 1
````

## Create Zip

````python
import zipfile
import os

def zip_directory(folder_path, zip_name):
    with zipfile.ZipFile(zip_name, 'w') as zip_file:
        for foldername, subfolders, filenames in os.walk(folder_path):
            for filename in filenames:
                file_path = os.path.join(foldername, filename)
                zip_file.write(file_path, os.path.relpath(file_path, folder_path))

zip_directory("out/", "podcast.zip")
````
